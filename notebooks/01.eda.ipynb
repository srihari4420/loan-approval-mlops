### Finding 1: Severe class imbalance

- Only 8.07% of applicants default (24,825 of 307,511).
- Implication: accuracy is meaningless — a model predicting "always repay" gets 91.93% accuracy.
- Must use ROC-AUC and Precision-Recall as primary metrics.
- Must use `stratify=y` in train/test split, and `StratifiedKFold` for CV.
- Will use `scale_pos_weight` in XGBoost (~11.4) rather than SMOTE — gradient boosting handles imbalance well natively.

### Finding 2: Significant missing data

- 49 columns have >50% missing values, mostly housing-related (COMMONAREA, NONLIVINGAPARTMENTS, etc.).
- 64 columns have >20% missing.
- Strategy:
  - Drop columns with >60% missing (likely too sparse to be useful).
  - For remaining numeric columns: impute with median.
  - For categorical columns: treat 'missing' as its own category (banking convention — sometimes missingness is informative).
  - Add binary `_was_missing` flag for top-10 most-missing-but-kept columns (the act of missingness can itself be predictive).

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print(f"Numeric columns:     {len(numeric_cols)}")
print(f"Categorical columns: {len(categorical_cols)}")
print(f"\nCategorical columns:")
for col in categorical_cols:
    n_unique = df[col].nunique()
    print(f"  {col}: {n_unique} unique values")
```

### Finding 4: Sentinel values masquerading as real data

- `DAYS_EMPLOYED = 365243` appears 55,374 times — clearly a "not applicable" placeholder.
- Counterintuitively, applicants with this sentinel have a **lower** default rate (5.4%) than employed applicants (8.6%).
- Likely represents pensioners or applicants whose employment isn't tracked — informative signal.
- Strategy: replace 365243 with NaN, add `DAYS_EMPLOYED_ANOM` binary flag indicating the sentinel.

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder

# Quick prep — encode categoricals as integers, fill NaN with -1
df_mi = df.copy()
for col in categorical_cols:
    df_mi[col] = LabelEncoder().fit_transform(df_mi[col].fillna('missing'))
df_mi = df_mi.fillna(-1)

X_mi = df_mi.drop(columns=['SK_ID_CURR', 'TARGET'])
y_mi = df_mi['TARGET']

# Sample for speed — 50k rows is plenty
sample_idx = np.random.RandomState(42).choice(len(X_mi), size=50000, replace=False)
mi_scores = mutual_info_classif(
    X_mi.iloc[sample_idx], 
    y_mi.iloc[sample_idx], 
    random_state=42,
    n_neighbors=3
)

mi_df = pd.DataFrame({
    'feature': X_mi.columns,
    'mi_score': mi_scores
}).sort_values('mi_score', ascending=False)

print("Top 20 features by mutual information:")
print(mi_df.head(20).to_string(index=False))
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
top20 = mi_df.head(20)
ax.barh(top20['feature'][::-1], top20['mi_score'][::-1])
ax.set_xlabel('Mutual Information')
ax.set_title('Top 20 features by predictive signal')
plt.tight_layout()
plt.show()
```

### Finding 5: Three features dominate predictive signal

- `EXT_SOURCE_2`, `EXT_SOURCE_3`, `EXT_SOURCE_1` are external credit bureau scores.
- Together they carry more signal than the next 50 features combined.
- Implication: a baseline using only these three already gets meaningful AUC. The remaining features add marginal gains.
- For the model: ensure these are never dropped during feature selection.
- For interpretability: SHAP values will be dominated by EXT_SOURCE — important for the explainability story.

```python
top_features = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, feat in zip(axes, top_features):
    for target_val in [0, 1]:
        subset = df[df['TARGET'] == target_val][feat].dropna()
        ax.hist(subset, bins=50, alpha=0.5, label=f'TARGET={target_val}', density=True)
    ax.set_title(feat)
    ax.set_xlabel(feat)
    ax.set_ylabel('Density')
    ax.legend()
plt.tight_layout()
plt.show()
```

# Home Credit Default Risk — EDA

**Goal:** Understand the dataset before building a model, so feature engineering and model choice are data-driven.

## Key findings

1. **Severe class imbalance** (8.07% defaults) — drives metric choice (ROC-AUC, not accuracy) and stratified splitting.
2. **49 columns with >50% missing values** — drop most, impute rest, add missingness flags.
3. **122 features mixed types** — 106 numeric, 16 categorical; mix of one-hot and target encoding needed.
4. **Sentinel values** in `DAYS_EMPLOYED` (365243) — must be cleaned and flagged.
5. **3 features dominate** — `EXT_SOURCE_1/2/3` carry the bulk of predictive signal.
6. **Default rate among "anomaly" applicants is lower** — missingness itself is informative.

## Implications for modelling

- Use ROC-AUC and Precision-Recall AUC as primary metrics.
- Baseline: logistic regression with median imputation, one-hot for low-card categoricals.
- Production model: XGBoost with `scale_pos_weight=11.4`, target encoding for `ORGANIZATION_TYPE`.
- Explainability: SHAP values will likely centre on `EXT_SOURCE_*` features.
- Feature count after engineering: aim for ~80 features (drop high-missing, add 5–10 flags and ratios).
